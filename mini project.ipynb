```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

fig, axes = plt.subplots(3, 2, figsize=(16, 14))
fig.suptitle("Simple Data Visualizations", fontsize=18)

# 1. Bar Plot: Average Salary per Years of Experience
avg_salary_per_exp = df.groupby("Years of Experience")["Salary ($)"].mean().reset_index()
sns.barplot(x="Years of Experience", y="Salary ($)", data=avg_salary_per_exp, ax=axes[0, 0])
axes[0, 0].set_title("Average Salary by Years of Experience")
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Line Plot: Salary over Index
sns.lineplot(x=df.index, y="Salary ($)", data=df, ax=axes[0, 1])
axes[0, 1].set_title("Salary Over Index")

# 3. Histogram: Distribution of Years of Experience
sns.histplot(df["Years of Experience"], bins=15, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Distribution of Years of Experience")

# 4. Scatter Plot: Experience vs. Salary
sns.scatterplot(x="Years of Experience", y="Salary ($)", data=df, ax=axes[1, 1])
axes[1, 1].set_title("Experience vs. Salary")

# 5. Box Plot: Salary Distribution
sns.boxplot(y="Salary ($)", data=df, ax=axes[2, 0])
axes[2, 0].set_title("Box Plot of Salary")

# Remove the empty subplot (bottom right)
fig.delaxes(axes[2, 1])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
